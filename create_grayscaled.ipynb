```python
def pblock(r, g, b):
    return fg(r, g, b) + '█' + fg.rs

def pspectrum(clist):
    for ctup in clist:
        print(pblock(*ctup), end="")
```

```python
def grayscale(clist):
    grayscaled = []
    for r, g, b in clist:
        gsvalue = int((r + g + b) / 3)
        grayscaled.append((gsvalue, gsvalue, gsvalue))

    return grayscaled

def gs(r, g, b):
    gsvalue = int((r + g + b) / 3 / 1.7) 
    return (gsvalue, gsvalue, gsvalue)
```

```python
def rgbcss(r, g, b):
    return f'#{hex(r)[2:].zfill(2)}{hex(g)[2:].zfill(2)}{hex(b)[2:].zfill(2)}'

def rgbacss(r, g, b):
    return f'rgba({r}, {g}, {b}, '
```

```python
def diff_clists(clista, clistb):
    for (r1, g1, b1), (r2, g2, b2) in zip(clista, clistb):
        print(f'{pblock(r1, g1, b1)} -> {pblock(r2, g2, b2)}: s/{rgbcss(r1, g1, b1)}/{rgbcss(r2, g2, b2)}/g; s/{rgbacss(r1, g1, b1)}/{rgbacss(r2, g2, b2)}/g;')

diff_clists(adapta, grayscale(adapta))
```
